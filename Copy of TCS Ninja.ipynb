###check whether the given number is prime or not
###program should accept a positive integer
###ifthe number is prime then print 'prime' or else print 'not prime'

```
n=int(input("enter a number"))
if n>=0:
  i=2
  count=0
  while(i<=n):
    if(n%i==0):
      count+=1
  
if count==0:
  print("prime")
else:
  print("not prime")
```
